#### Exercise 1
Adapted from [TCS] Appendix D, Exercise 2

Open help for the math module.

- What are the constants in the math module?
  
  Answer: math.pi for pi, math.e for logarithm, math.tau for the 2pi
- What's does `math.exp` do? What about `math.pow`?  

  Answer: math.exp returns the value of x raised to the power of y the same as math.pow. The only difference is that math.exp accepts 1 parameter and math.pow accepts 2 parameter
- How do you calculate logarithms? Logarithms base 10? Natural Logarithms?

  Answer: Logarithm base 10 -> variable_name = math.log10(x) natural logarithm -> variable_name = math.log(x) use import math module
