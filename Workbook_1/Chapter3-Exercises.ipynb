#### Exercise 1

In the code below, which will never print regardless of the value for x? Edit the code to correct the error.

```python
if x > 0 :
    print('Above 0')
elif x >  10 :
    print('Above 10')
elif x > 50 :
    print('Above 50')
elif x >  20 :
    print('Above 20')
else :
    print('Something else')
```


Answer: The condition x > 10 : will never be checked, because if x > 0 is true, it means x is a positive number, and the next conditions x > 10, x > 50 and x > 20 should be checked in that order. The corrected code should look like this:

```python
if x > 0 :
    print('Above 0')
elif x > 10 :
    print('Above 10')
elif x > 20 :
    print('Above 20')
elif x > 50 :
    print('Above 50')
else :
    print('Something else')
```

#### Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

````python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))
pay = hours * rate
if pay < 40 :
    print(pay)
else :
    print("Pay: " + str(((hours-40)*(1.5*rate) + (40*rate))))
````

## CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

Adapted from [PEW] E3.3
<br>

````python
score = float(input("Enter the score (0-100): "))

if score >= 97:
    grade = 1.0
elif score >= 93:
    grade = 1.25
elif score >= 89:
    grade = 1.5
elif score >= 85:
    grade = 1.75
elif score >= 81:
    grade = 2.0
elif score >= 77:
    grade = 2.25
elif score >= 74:
    grade = 2.5
else:
    grade = 3.0

print("Grade:", grade)
````
