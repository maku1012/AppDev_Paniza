## EXERCISES

#### Exercise 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E4.6

```python
def computepay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        pay = (40 * rate) + ((hours - 40) * 1.5 * rate)
    return pay

def main():
    hours = float(input("Enter Hours: "))
    rate = float(input("Enter Rate: "))
    pay = computepay(hours, rate)
    print("Pay:", pay)

main()
```

## CHALLENGE

#### Challenge

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

    Run the program repeatedly to test the various different values for input.
    
Adapted from [PEW] E4.6
<br>

```python
score = float(input("Enter the score (0-100): "))

def computegrade(score):
    if score >= 97:
        grade = 1.0
    elif score >= 93:
        grade = 1.25
    elif score >= 89:
        grade = 1.5
    elif score >= 85:
        grade = 1.75
    elif score >= 81:
        grade = 2.0
    elif score >= 77:
        grade = 2.25
    elif score >= 74:
        grade = 2.5
    else:
        grade = 3.0
    return grade

while True:
    grade = computegrade(score)
    print("Grade:", grade)
    continue_or_exit = input("continue? (y/n): ")
    if continue_or_exit.lower() != "y":
        break
```
