#### Exercise 1

Given the code below    

    text = 'Business Analytics'

What will the following expressions produce?
    
    text[4]
    text[2:4]
    text[:4]
    text[4:]
    text[-2]
    text[-4:-2]
    text[-2:]
    text[:-2]
    

```python
text = 'Business Analytics'
print(text[4])
print(text[2:4])
print(text[:4])
print(text[4:])
print(text[-2])
print(text[-4:-2])
print(text[-2:])
print(text[:-2])
```


## CHALLENGE


Write a program which repeatedly reads a **full name**, then does the following:
- strips the name of any trailing spaces
- capitalizes the first letter
- puts the rest into smaller letters


```python
def format_name(full_name):
    formatted_name = full_name.strip().capitalize()
    formatted_name = formatted_name[:1] + formatted_name[1:].lower()
    return formatted_name
def main():
    while True:
        full_name = input("Enter a full name (or 'quit' to exit): ")
        if full_name.lower() == 'quit':
            print("Exiting program...")
            break
        formatted_name = format_name(full_name)
        print("Formatted name:", formatted_name)
if __name__ == "__main__":
    main()
```
